```python

import matplotlib.pyplot as plt
import numpy as np

# 
plt.rcParams['font.sans-serif'] = ['SimHei'] #
plt.rcParams['axes.unicode_minus'] = False #

# 
time = np.arange(1, 13)  # 1 - 12

# %
up_down_enterprises = [39.8, 39.8, 39.8, 39.8, 39, 39, 38, 38, 37, 38, 39.8, 39.8]

# %
speculative_capital = [35, 33, 30, 28, 32, 34, 36, 34, 32, 30, 31, 33]

# %
financial_institutions = [34.5, 33, 28, 32, 34, 36, 34, 32, 30, 31, 33, 32.5]

# /
price = [10, 12, 15, 18, 16, 14, 13, 14, 17, 16, 15, 14]

# 
plt.figure(figsize=(12, 6))

# 
line1, = plt.plot(time, up_down_enterprises, label='', marker='^', color='green')

# 
line2, = plt.plot(time, speculative_capital, label='', marker='o', color='blue')

# 
line3, = plt.plot(time, financial_institutions, label='', marker='s', color='orange')

# Y
ax2 = plt.twinx()
line4, = ax2.plot(time, price, label='', color='red', marker='d')
ax2.set_ylabel('/', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Y
plt.ylabel('%')

# X
plt.xlabel('')
plt.title('')

# 
lines = [line1, line2, line3, line4]
labels = [line.get_label() for line in lines]
plt.legend(lines, labels, loc='upper right')

# 
plt.grid(True, linestyle='--', alpha=0.7)

# 
plt.show()
```
