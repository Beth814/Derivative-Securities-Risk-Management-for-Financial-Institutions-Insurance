```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei'] #设置能显示中文
plt.rcParams['axes.unicode_minus'] = False #设置负号


# 生成 6 个月期 LIBOR 数据
libor = np.linspace(5.0, 12.0, 100)

# 浮动利率筹资成本：LIBOR + 0.5%
floating_cost = libor + 0.5

# 零成本封顶保底式筹资成本
zero_cost = np.where(libor <= 7.5, (7.5 + 0.5) - (7.5 - libor) * 0.35, 8.0)

# 绘制图形
plt.figure(figsize=(8, 6))
plt.plot(libor, floating_cost, label='浮动利率筹资成本', linestyle='-')
plt.plot(libor, zero_cost, label='零成本封顶保底交易筹资成本', linestyle='-')

# 标注特殊点（7.5, 8.0）
plt.plot(7.5, 8.0, 'ro')
plt.vlines(7.5, 5.0, 8.0, linestyles='dashed')
plt.hlines(8.0, 5.0, 7.5, linestyles='dashed')

# 添加箭头标注
plt.annotate('封顶协定价格点', xy=(7.5, 8.0), xytext=(8, 7.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='left', verticalalignment='bottom')
plt.annotate('浮动成本线', xy=(10, 10.5), xytext=(10.5, 10),
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='left', verticalalignment='bottom')
plt.annotate('封顶后成本线', xy=(10, 8), xytext=(10.5, 7.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='left', verticalalignment='bottom')

# 设置坐标轴标签和标题
plt.xlabel('6 个月期的 LIBOR')
plt.ylabel('实际筹资成本(%)')
plt.title('零成本封顶保底式利率期权交易筹资成本')
plt.legend()

# 显示网格
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图形
plt.show()
    
```
