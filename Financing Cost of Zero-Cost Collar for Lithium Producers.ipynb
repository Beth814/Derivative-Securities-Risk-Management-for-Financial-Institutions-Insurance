```python
import numpy as np
import matplotlib.pyplot as plt

# 
plt.rcParams['font.sans-serif'] = ['SimHei'] #
plt.rcParams['axes.unicode_minus'] = False #


#  6  LIBOR 
libor = np.linspace(5.0, 12.0, 100)

# LIBOR + 0.5%
floating_cost = libor + 0.5

# 
zero_cost = np.where(libor <= 7.5, (7.5 + 0.5) - (7.5 - libor) * 0.35, 8.0)

# 
plt.figure(figsize=(8, 6))
plt.plot(libor, floating_cost, label='', linestyle='-')
plt.plot(libor, zero_cost, label='', linestyle='-')

# 7.5, 8.0
plt.plot(7.5, 8.0, 'ro')
plt.vlines(7.5, 5.0, 8.0, linestyles='dashed')
plt.hlines(8.0, 5.0, 7.5, linestyles='dashed')

# 
plt.annotate('', xy=(7.5, 8.0), xytext=(8, 7.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='left', verticalalignment='bottom')
plt.annotate('', xy=(10, 10.5), xytext=(10.5, 10),
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='left', verticalalignment='bottom')
plt.annotate('', xy=(10, 8), xytext=(10.5, 7.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='left', verticalalignment='bottom')

# 
plt.xlabel('6  LIBOR')
plt.ylabel('(%)')
plt.title('')
plt.legend()

# 
plt.grid(True, linestyle='--', alpha=0.7)

# 
plt.show()
    
```
