```python
import matplotlib.pyplot as plt
import numpy as np

# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial']  # 添加英文字体备选
plt.rcParams['axes.unicode_minus'] = False

# 模拟数据（假设负相关性-0.65）
np.random.seed(42)
x = np.arange(100)  # APAC-PDI指数（0-100点）
y = -0.65 * x + np.random.normal(0, 10, 100)  # 私募债组合净值

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.6)
plt.plot(x, -0.65*x + 50, color='red', linestyle='--', label='Trend Line (Corr=-0.65)')

plt.title("Correlation between APAC-PDI Index and Private Debt Portfolio NAV (2020-2023)", fontsize=14)
plt.xlabel("APAC-PDI Index (Points)", fontsize=12)
plt.ylabel("Private Debt Portfolio NAV (AUD Million)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.text(70, -30, "Data Source: S&P Global", fontsize=10, color='gray')
plt.show()    
```
